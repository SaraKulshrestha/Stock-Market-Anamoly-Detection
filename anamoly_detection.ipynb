```python
import pandas as pd
import yfinance as yf 
from datetime import date , timedelta
```

```python
# defining start date as exactly one year ago and end date as current date
# analysing two year
endDate = date.today().strftime("%Y-%m-%d")
startDate = (date.today() - timedelta(days=1200)).strftime("%Y-%m-%d")
```

```python
# stock tickers (Tickers are unique symbols or abbreviations used to identify publicly traded companies on stock exchanges)
tickers = ['AAPL', 'MSFT', 'GOOG','TSLA','NVDA','PEP','JPM','GS']
```

```python
data = yf.download(tickers, start=startDate, end=endDate, progress=True)
```

```python
data = data.reset_index()
```

```python
data
```

```python
# function in Pandas is used to "unpivot" a DataFrame from a wide format to a long format
# id_vars specifies the columns that should remain unchanged.
# var_name specifies the names of the new columns created by the melt operation.


data_melted = data.melt(id_vars=['Date'], var_name=['Attribute', 'Ticker'])
```

```python
data
```

```python
#reshapes the long-format DataFrame back into a wide format but organizes it differently.
#pivots the data_melted DataFrame to make Attribute values (like Open, High, Low, etc.) the columns for each Date and Ticker.
#aggfunc='first': If there are multiple entries for the same Date, Ticker, and Attribute, this will take the first occurrence. 


pivoted = data_melted.pivot_table(index=['Date', 'Ticker'], columns='Attribute', values='value', aggfunc='first')
```

```python
#converts the Date and Ticker index back into regular columns.
#

stock = pivoted.reset_index()
```

```python
stock.head(5)
```

```python
# Date: The date of the stock data entry.
# Ticker: The stock ticker symbol.
# Adj Close: The adjusted closing price of the stock, which accounts for any corporate actions like splits or dividends.
# Close: The closing price of the stock.
# High: The highest price of the stock during the trading day.
# Low: The lowest price of the stock during the trading day.
# Open: The opening price of the stock.
# Volume: The number of shares traded during the day.
```

```python
stock["Date"] = pd.to_datetime(stock["Date"]) #datetime format conversion

stock.set_index("Date",inplace =True) #date as index
stock.head(5)
```

```python
# anamoly detection can focus on---
# Significant price movements that deviate from the stock’s typical price range or trend.
# Unusual trading volumes that stand out from the normal trading activity.
```

```python
# A common threshold for identifying an anomaly is a Z-score greater than 2 or less than -2,
# which corresponds to data points that are more than 2 standard deviations away from the mean.
```

```python
from scipy.stats import zscore
```

```python
def detect_anomalies(df, column):
    df2 = df.copy()
    df2["Z_score"] = zscore(df2[column])
    anamolies = df2[abs(df2["Z_score"]) > 2] #absolute Z-score is greater than 2 (common threshold for anomalies)
    return anamolies
```

```python
aac = pd.DataFrame()
av = pd.DataFrame()
```

```python
for ticker in stock['Ticker'].unique():
    dataTicker = stock[stock['Ticker'] == ticker]

    adj_close_anomalies = detect_anomalies(dataTicker, 'Adj Close')
    volume_anomalies = detect_anomalies(dataTicker, 'Volume')
```

```python
aac = pd.concat([aac, adj_close_anomalies])
av = pd.concat([av, volume_anomalies])
```

```python
def plot_anamolies(ticker,aac,av):
    dataTicker = stock[stock["Ticker"] == ticker]
    volume_anomalies = av[av["Ticker"] == ticker]
    adj_close_anomalies = aac[aac["Ticker"] == ticker]
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 12))
    
    
# Adjusted Closing Price
    ax1.plot(dataTicker.index, dataTicker['Adj Close'], label='Adj Close', color='pink')
    ax1.scatter(adj_close_anomalies.index, adj_close_anomalies['Adj Close'], color='red', label='Anomalies')
    ax1.set_title(f'{ticker} Adjusted Close Price and Anomalies')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Adjusted Close Price')
    ax1.legend()
    
#VOLUME
    ax2.plot(dataTicker.index, dataTicker['Volume'], label='Volume', color='purple')
    ax2.scatter(volume_anomalies.index, volume_anomalies['Volume'], color='blue', label='Anomalies')
    ax2.set_title(f'{ticker} Trading Volume and Anomalies')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('Volume')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
# plotting for every ticker
for ticker in stock['Ticker'].unique():
    plot_anamolies(ticker, aac, av)
```

```python
all_aac = aac[['Ticker']].copy()
all_aac['Adj Close Anomaly'] = 1 #1 for anamolies
```

```python
all_av = av[['Ticker']].copy()
all_av['Volume Anomaly'] = 1 
```

```python
adj_close_pivot = all_aac.pivot_table(index=all_aac.index, columns='Ticker',
                                                       fill_value=0, aggfunc='sum')

# aggfunc='sum' means that if there are multiple entries for the same date and ticker, their values will be summed
# Rows: Each row corresponds to a unique date.
# Columns: Each column corresponds to a unique ticker.
# Values: The values in the DataFrame are the sum of adjusted closing prices for each ticker on each date. Missing values are filled with 0.
```

```python
volume_pivot = all_av.pivot_table(index=all_av.index, columns='Ticker',
                                                fill_value=0, aggfunc='sum')
```

```python
adj_close_pivot
```

```python
# Check the number of levels in the index
if adj_close_pivot.columns.nlevels > 1:
    adj_close_pivot.columns = adj_close_pivot.columns.get_level_values(1)

if volume_pivot.columns.nlevels > 1:
    volume_pivot.columns = volume_pivot.columns.get_level_values(1)
```

```python
# combine the two pivoted dataframes

combined_anomalies = pd.concat([adj_close_pivot, volume_pivot], axis=1, keys=['Adj Close Anomaly', 'Volume Anomaly'])
```

```python
# calculate the correlation matrix for the anomalies
correlation_matrix = combined_anomalies.corr()
```

```python
# Frequency of Anomalies: A higher number of anomalies may indicate a higher risk.
# Magnitude of Anomalies: Larger deviations from the mean (higher absolute Z-scores) suggest higher risk.
```

```python
#rating each stock based on the risk inferred from the anomalies detected
#A stock could be considered more risky if it has frequent and large anomalies in its price or volume

# Frequency of Anomalies: A higher number of anomalies may indicate a higher risk.
# Magnitude of Anomalies: Larger deviations from the mean (higher absolute Z-scores) suggest higher risk.

#computing a risk score for each stock by combining these factors
#average the absolute Z-scores of anomalies for each stock and then normalize these scores across all stocks to get a risk rating:


# calculate the mean absolute Z-score for each stock as a risk indicator
adj_close_risk = aac.groupby('Ticker')['Z_score'].apply(lambda x: abs(x).mean())
volume_risk = av.groupby('Ticker')['Z_score'].apply(lambda x: abs(x).mean())

# combine the risk scores from both price and volume anomalies
total_risk = adj_close_risk + volume_risk

# normalize the risk scores to get a relative risk rating from 0 to 1
risk_rating = (total_risk - total_risk.min()) / (total_risk.max() - total_risk.min())

print(risk_rating)
```
